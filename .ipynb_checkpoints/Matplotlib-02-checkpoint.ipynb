# Matplotlib 02: Plot Times Series Data

Times series data is one of the most common data which appeares in may fields such as physics, astonomy, finance, biology and etc.

Python's `pandas` library and `matplotlib` together give us tools to easily plot time series data. To get benefit of Matplotlib's times series data visualization, we will used pandas to read data, parse dates and set the index of Pandas dataframe to a datatime object.

```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
```

## Read a times series data with pandas

First, Let us read **global temperature** data which is a times series data of monthly averaged temperture of the earth.

To read more about this dataset click [here](https://datahub.io/core/global-temp#readme).

```python
global_warming_df = pd.read_csv("./data/global-warming.csv")
```

```python
global_warming_df.head()
```

As we see the column with header named `Date` is a date column. We tell Pandas to consider this column as a Pandas datetime object using `parset_dates=True` and `index_col='Date'` in `pd.read_csv()`.

```python
global_warming_df = pd.read_csv("./data/global-warming.csv", parse_dates=True, index_col='date')
global_warming_df.sort_index(inplace=True)
```

```python
global_warming_df.index
```

Now, we can use Matplotlib to plot the times series of global temperature.

## Matplotlib Times Series

In this section, we use Matplotlib to plot time series data.

```python
# example 01: plot time series
fig, ax = plt.subplots()

ax.plot(global_warming_df.index, global_warming_df["mean temperature"])
ax.set_title('Global mean temperature deviations from the corresponding 1951-1980 means')
ax.set_xlabel("Date")
ax.set_ylabel("Temperature mean deviation (Celsius degree)")

plt.show()
```

As you see Matplotlib automatically handled the x-axis and extracted years to set them as ticks of the axis.
Now, lets choose a smaller interval to plot and see how Matplotlib handels that.

```python
# example 01: plot time series
fig, ax = plt.subplots()

dataset = global_warming_df["2015-01-01":"2016-01-01"]

ax.plot(dataset.index, dataset["mean temperature"])
ax.set_title('Global mean temperature deviations from the corresponding 1951-1980 means')
ax.set_xlabel("Date")
ax.set_ylabel("Temperature mean deviation (Celsius degree)")

plt.show()
```

As you see, Matplotlib automatically changes the ticks of x-axis from year to month when the times series x-axis ranges gets smaller than a year

## Plot two times series in one subplot

Sometimes it may be useful two plot two time series in same subplot where the time index of two variables is same. In this section, we are going to explain how to do that.

To plot two times series which have same times index, i.e. same x-axis but different y-axis we use `ax.twinx()` method of Matplotlib.

```python
# example 02: plot two times series
fig, ax = plt.subplots()

# plot the first times series
ax.plot(global_warming_df.index, global_warming_df['mean temperature'])

# add the second plot in the subplot
ax2 = ax.twinx()
ax2.plot(global_warming_df.index, global_warming_df['co2 (ppm)'])

plt.show()
```

As we see the two sime series are plotted in same subplot. Note that the left and right y-axis are different. In addition, in this section, we select data which are after **'1958-01-01'**.

```python
# example 02: plot two times series
fig, ax = plt.subplots(figsize=(12,6))

# select dataset
dataset = global_warming_df['1958-01-01':]

# plot the first times series
ax.plot(
    dataset.index,
    dataset['mean temperature'],
    color='blue',
)
ax.set_ylabel('Temperature mean deviation (Celsius degree)', color='blue')
ax.tick_params('y', colors='blue')
ax.set_title("Global CO2 and mean temperature")


# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()
# add the second plot in the subplot
ax2.plot(
    dataset.index,
    dataset['co2 (ppm)'],
    color='red'
)
ax2.set_ylabel('CO2 (ppm)', color='red')
ax2.tick_params('y', colors='red')

plt.show()
```

## Annotations

Sometimes, in plots it is useful to point to some specific information which plot conveys. In this kind of situations, the **annotations** is what wee need to add to the plot. To add annotations for the plot we use `ax.annotate()` with appropriate parameters as follows:

```python
# example 03: plot annotations
fig, ax = plt.subplots(figsize=(12,6))

# select dataset
dataset = global_warming_df['1958-01-01':]

# plot the first times series
ax.plot(
    dataset.index,
    dataset['mean temperature'],
    color='blue',
)
ax.set_ylabel('Temperature mean deviation (Celsius degree)', color='blue')
ax.tick_params('y', colors='blue')
ax.set_title("Global CO2 and mean temperature")


# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()
# add the second plot in the subplot
ax2.plot(
    dataset.index,
    dataset['co2 (ppm)'],
    color='red'
)
ax2.set_ylabel('CO2 (ppm)', color='red')
ax2.tick_params('y', colors='red')


# add annotations
ax.annotate(
    '> 1 Celcius',
    xy=[pd.Timestamp('2016-02-06'),1.35],
    xytext=[pd.Timestamp('2000-02-06'),1.35],
    arrowprops = {"arrowstyle":'->', "color":'green'},
)

plt.show()
```
