## Matplotlib "Hello World!"

The first program, we want to write in matplotlib is to just create an empty graph. This plot is the "Hello World!" program of the matplotlib.

````python
# example 1: empty figure

fig, ax = plt.subplots()  # create a figure and a set of subplots

plt.show()  # show the plot
````

The following command
```python
fig, ax = plt.subplots()
```
creates two objects. The ```fig``` is an object of type `matplotlib.figure.Figure`. The ```ax``` objects is an object of type `matplotlib.axes._subplots.AxesSubplot`.


We can set parameter `'figsize'` to change the size of the plot.

````python
# example 1: empty figure

fig, ax = plt.subplots(figsize=(10,2))  # create a figure and a set of subplots

plt.show()  # show the plot
````

## First plot

Let's do our first visualization with matplotlib. We read `iran-population.csv` dataset which contains the population of Iran's people. Let's read the data with pandas and see what it looks like:

````python
iran_population_df = pd.read_csv("data/iran-population.csv")
iran_population_df
````

### One graph

Now, let use matplotlib to plot the total population of Iran people vs year.

````python
# example 2: one plot in a canvas

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['total']
)

plt.show()
````

### Two graphs
In previous example, we add on plot in our figure. Now, we want to add __two figures__. We plot the **urban** and **rural** population of Iran in a same canvas.

````python
# example 3: two plots in a canvas

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['urban']
)
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['rural']
)

plt.show()
````

## Customizing plots

### Add labels to plot

When you plot a graph, ALWAYS remeber to add labels to axes. In the following example, we add labels to plot.

````python
# example 4: two plots in a canvas with labels

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['urban']
)
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['rural']
)

# set x axis label
ax.set_xlabel('Year (jalali)')
# set y axis label
ax.set_ylabel('Number of population')
# set title for the plot
ax.set_title('Population of Iran vs jalali years')

plt.show()
````

In addition, we need to label each graph in the plot if there is more than one. In matplotlib, we use `ax.legend()` to show the labels of each graph.

````python
# example 4.1: two plots in a canvas with labels

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['urban'],
    label='urban',
)
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['rural'],
    label='rural',
)

# set x axis label
ax.set_xlabel('Year (jalali)')
# set y axis label
ax.set_ylabel('Number of population')
# set title for the plot
ax.set_title('Population of Iran vs jalali years')

# add legends
ax.legend()

plt.show()
````

### Add markers, color and linestyle

The figure above is a bit misleading. The population survey has done every 10 years, but the data is a continuous line which may communicate this message that the population is surveyed every year. To eliminate this mistake, we can add **markers** in where we have actual data points. In addition, we can change the style of line to show that these lines are just ways to connect data points. Moreover, with matplotlib we can change the color of plots to whatever color we want.

````python
# example 4: two plots in a canvas with labels

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['urban'],
    marker='o',
    linestyle='dashed',
    color='purple',
    label='urban',
)
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['rural'],
    marker='^',
    linestyle='dashed',
    color='green',
    label='rural',
)

# set x axis label
ax.set_xlabel('Year (jalali)')
# set y axis label
ax.set_ylabel('Number of population')
# set title for the plot
ax.set_title('Population of Iran vs jalali years')
# add legend
ax.legend()

plt.show()
````

> To see the list of available **linestyles** in matplotlib look at [here]("https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html").

Matplot References:

````python
# example 4: two plots in a canvas with labels

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['urban'],
    marker='o',
    linestyle='None',
    color='purple',
    label='urban'
)
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['rural'],
    marker='^',
    linestyle='None',
    color='green',
)

# set x axis label
ax.set_xlabel('Year (jalali)')
# set y axis label
ax.set_ylabel('Number of population')
# set title for the plot
ax.set_title('Population of Iran vs jalali years')

plt.show()
````

## Small multiples

Small mulitples are plots which are drawn side-by-side. To have small multiples, we use Matplotlib's `plt.subplots()` function. The first argument is the number of rows (`nrows`) and the second is the number of columns (`ncols`). Let's create empty subplots.

````python
# example 5: empty plots in 2 rows ans 3 columns
fig, ax = plt.subplots(3,2)

plt.show()
````

Its a bt ugly, because canvases go through each other. To make the plot to look better, we use `plt.tight_layout()`

````python
# example 6: empty plots in 2 rows ans 3 columns
fig, ax = plt.subplots(3,2)

plt.tight_layout()
plt.show()
````

When you write the `fig, ax = plt.subplots(2,3)`, now the `ax` objects is not of type `matplotlib.axes._subplots.AxesSubplot`. The type of `ax` is now a numpy array (`numpy.ndarray`).

````python
type(ax)
````

Each element in `ax` is now an `matplotlib.axes._subplots.AxesSubplot` object. To access any of them, we can use numpy array slicing.

````python
ax
````

**Warning**: 
1. When `nrows > 1` and `ncols > 1` we must use `ax[i, j]` to access the axis in `i`th row and `j`th column.
2. When `nrows = 1` and `ncols > 1` (or `nrows > 1` and `ncols = 1`) we must us `ax[i]` to access the `i`th axis.
3. When `nrows = 1` and `ncols = 1`, we must use `ax` itself to access the (the only one) axis.

````python
ax[0,0]
````

Let's define a subplot and add some plot to each axis

````python
# example 7: plots in 2 rows and 2 columns
fig, ax = plt.subplots(2,2)

ax[0,0].plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
)

ax[0, 1].plot(
    iran_population_df['jyear'],
    iran_population_df['urban'],
)

ax[0, 1].plot(
    iran_population_df['jyear'],
    iran_population_df['rural'],
)

ax[1, 0].plot(
    iran_population_df['rural'],
    iran_population_df['urban'],
)

ax[1, 1].plot(
    iran_population_df['jyear'],
    iran_population_df['rural'] / iran_population_df['urban'],
)

plt.tight_layout()
plt.show()
````

When we want to compare subplots we can use `sharey` or `sharex` parameters as follows:

````python
# example 8: two plots with shared scale
fig, ax = plt.subplots(nrows=2, sharey=True)

ax[0].plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
    linestyle='solid',
    color='blue',
    marker='s'
)

ax[1].plot(
    iran_population_df['jyear'],
    iran_population_df['urban']-iran_population_df['rural'],
    linestyle='dashed',
    color='red',
    marker='o'
)


plt.tight_layout()
plt.show()
````

## Plot styles

Styles are some settings that will be applied in all plots such as backgroud color, grid, .... Matplotlib has some predefiend style which can be set using `plt.style.use()` function.

Let's change style of Matplotlib and plot the previous plot.

````python
# example 9: two plots with shared scale with ggplot2 style
# change plot style
plt.style.use('ggplot')

fig, ax = plt.subplots(nrows=2, sharey=True)

ax[0].plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
    linestyle='solid',
    color='blue',
    marker='s'
)

ax[1].plot(
    iran_population_df['jyear'],
    iran_population_df['urban']-iran_population_df['rural'],
    linestyle='dashed',
    color='red',
    marker='o'
)


plt.tight_layout()
plt.show()
````

To see the available styles execute `plt.style.available`

````python
plt.style.available
````

````python
# example 9: two plots with shared scale with ggplot2 style
# change plot style
plt.style.use('seaborn-colorblind')

fig, ax = plt.subplots(nrows=2, sharey=True)

ax[0].plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
    linestyle='solid',
    color='blue',
    marker='s'
)

ax[1].plot(
    iran_population_df['jyear'],
    iran_population_df['urban']-iran_population_df['rural'],
    linestyle='dashed',
    color='red',
    marker='o'
)


plt.tight_layout()
plt.show()
````

After you change the style using `plt.styl.use` all plots will use that style until you change the style again. To back to default style of the Matplotlib use `plt.style.use('default')`

````python
plt.style.use('default')
````

### Tips to consider  when choosing styles

1. Dark backrounds make the plot harder to read. Do not use them if not needed.
2. Darker backgrouns sometimes may be useful. For example, when displaying a dashboard over web application. Very white background bothers the eye of human.             
3. If color are important in your plot, use styles which are designed for color blineded people. In Matplotlib, `'seabor-colorblind'` or `'tableau-colorblind10'`.
4. If the user of your plot will need to print the plot with black and white printer, then use `'grayscale'` style.
5. Try to use link, especially when the user of the plot might be willing to print your plot.

## Save figure

To save figures in files, we use `fig.savefig()` method.   
To change the size of plot, we use `fig.set_size_inches([x,y])` where `x` is the width and `y` is the height of the figure.       
To save plots without loss we can save them as `.png` or `.svg` formats. The `.svg` format is especially useful when you need to edit your plot in softwares such as Adobe.      
The `.jpg` format causes to loss the quality of the image and it is useful in some cases. For example, if you want to show the image in the website.

The `dpi` paramter in `fig.savefig()` can be used to determine the quality and size of the image file. It is used for image resoultion. The larger the `dpi` the higher the quality of the image and the larger its size.

````python
# example 10: Save plot

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['total']
)

fig.savefig('iran_population.png', dpi=300)
````

````python
# example 11: Save plot

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
)

fig.savefig('iran_population.jpg', dpi=100)
````

````python
# example 11: Change size of the plot and save plot

fig, ax = plt.subplots()

# add the total population plot
ax.plot(
    iran_population_df['jyear'],
    iran_population_df['total'],
)

# change size of the plot to 10 inches widths and 2 inches height
fig.set_size_inches([10,2])
fig.savefig('iran_population.svg')
````
